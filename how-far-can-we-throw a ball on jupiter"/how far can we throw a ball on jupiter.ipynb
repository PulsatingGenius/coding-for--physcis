#  - Plotting trajectories on different planets

Now I want to plot a projectile launced at *45* degrees on each of the planets. 

Be careful when using the values of g , make sure you use the right sign in the right place.

```
# EDIT THIS ONE...
# I'm assuming you've already executed the previous cell...

import matplotlib.pyplot as plt

v_i = 100 
a_x = 0

angle = 60


# lets loop through each of your planets and pick g at each point, as opposed to 'angles[i]
for i in range(len(names)):
    
    # get the timestamps for each projectile:
    t_max = 2 * v_i * np.sin(np.deg2rad(angle)) / -g
    t_step = 0.01
    timestamps = np.arange(0, t_max, t_step)
    
    # we then work out the x/y coordinates of our projectile at each timestamp for that angle
    x = v_i * timestamps * np.cos(np.deg2rad(angle)) + (0.5 * a_x * timestamps ** 2)
    y = v_i * timestamps * np.sin(np.deg2rad(angle)) + (0.5 * g * timestamps ** 2)
 
    plt.plot(x, y, label = names[i])

plt.show()
```

# Q4 - Why are we doing this?

You might be wondering why anyone would bother doing this, when we already know the equations of motion and could easily have worked out these results by hand in just a few seconds...

Well, you could imagine that our primitave projectile model can become much more complicated if we started to increase the 'realism' of our simulation. We could add:

*   Air resistance
*   Bouncing
*   Additional forces at launch
*   2D or even 3D force vectors
*   Multiple projectiles
*   A complex surface for the projectile to encounter
*   Projectiles that aren't point particles, e.g. something that has a shape
*   Material properties of the projectile/bouncing etc

And before you know it, we could be simulating:

*   A multistage rocket en-route to Mars
*   An explosive gas reaching thousands of degrees
*   Complex astrophysics of objects falling into black holes

**Q1** Printing various accelerations due to gravity:


```
import numpy as np

# planet mass in kg
# planet radius in m
# you can just use the 'e' notation: a*(10**b) == a e b

masses = np.array((3.285e23, 5.972e24, 6.39e23, 1.898e27))
radii  = np.array((2.440e6, 6.371e6, 3.390e6, 69.911e6))
G = 6.674e-11
names = ['Mercury', 'Earth', 'Mars', 'Jupiter']

g = (-G * masses / (radii**2))

for i in range(len(masses)):
    print ('Acceleration due to gravity at the surface of {a} is {b} m/s^2'.format(a=names[i], b=g[i]))
```

**Q2** Plotting various trajectories on different planets:

```
import matplotlib.pyplot as plt

v_i = 100 
a_x = 0

angle = 45


# lets loop through each of those angle values and work out the trajectory
for i in range(len(masses)):
    
    # each projectile gets a unique set of timestamps, so that each one is sampled 100 times
    t_max = 2 * v_i * np.sin(np.deg2rad(angle)) / -g[i]
    t_step = 0.01
    timestamps = np.arange(0, t_max, t_step)
    
    # we then work out the x/y coordinates of our projectile at each timestamp for that angle
    x = v_i * timestamps * np.cos(np.deg2rad(angle)) + (0.5 * a_x * timestamps ** 2)
    y = v_i * timestamps * np.sin(np.deg2rad(angle)) + (0.5 * g[i] * timestamps ** 2)
    
    #print (t_max)
    # add that data to the plot, give it a label
    plt.plot(x, y, label = names[i])


plt.legend()
plt.xlabel('Distance / m')
plt.ylabel('Height / m')
plt.title('The trajectory of a particle launched on various planets')
plt.show()
```

**Q3**. Plotting a realistic trajectory and printing the maximum distance on each planet.

```
import matplotlib.pyplot as plt

v_i = 20 
a_x = 0

angle = 45

max_distance = []


# lets loop through each of those angle values and work out the trajectory
for i in range(len(masses)):
    
    # each projectile gets a unique set of timestamps, so that each one is sampled 100 times
    t_max = 2 * v_i * np.sin(np.deg2rad(angle)) / -g[i]
    t_step = 0.01
    timestamps = np.arange(0, t_max, t_step)
    
    # we then work out the x/y coordinates of our projectile at each timestamp for that angle
    x = v_i * timestamps * np.cos(np.deg2rad(angle)) + (0.5 * a_x * timestamps ** 2)
    y = v_i * timestamps * np.sin(np.deg2rad(angle)) + (0.5 * g[i] * timestamps ** 2)
    
    max_distance.append(x[-1])
    
    #print (t_max)
    # add that data to the plot, give it a label
    plt.plot(x, y, label = names[i])


plt.legend()
plt.xlabel('Distance / m')
plt.ylabel('Height / m')
plt.title('The trajectory of a particle launched on various planets')
plt.show()

for i in range(len(names)):
    print ('Projectile launched from planet {a} reached {b} meters'.format(a=names[i], b=(np.round(max_distance[i],2))))
```
